Q1. Create a CSV file for address book. It should have columns for name, address, mobile, email. Insert 2-3 dummy data entered by user

```python
import csv
file=open("address_book.csv","w",newline="")
write=csv.writer(file)
write.writerow(["name","address","mobile","email"])

for i in range(3):
    print(f"enter details of person: {i}")
    name=input("enter name")
    address=input("enter address")
    mobile=input("enter mobile number")
    email=input("enter email")

    write.writerow([name, address,mobile,email])
file.close()
```

Q2. a.Create Database b.Create 2-3 tables

```python
import sqlite3
conn=sqlite3.connect("student.db")
conn.execute("""
CREATE TABLE student(id int PRIMARY KEY, name char(10), course char(10))
""")
conn.execute("""
CREATE TABLE marks(id int, subject char(10), marks float)
""")

conn.execute("""
CREATE TABLE teacher(tid int, tname char(10), department char(10))
""")

conn.commit()
conn.close()
```

d. Perform different select oprations
