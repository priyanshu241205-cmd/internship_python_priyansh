1) Practise Panda Series
a.Create a Panda Series from Dictionary

```python
# Panda series from dictionary
import pandas as pd
d={
    "subject":["python","Java","R","C++"],
    "marks":[92,93,90,88]
}

series=pd.Series(d)
print(d)

print(series.index)
print(series["marks"][0])  # accessing an element in the series

# converting series to lists
print(series.tolist)
```

b.Create Pandas Series using lists
c.Accessing elements of a series in Pandas

```python
# pandas series using lists
import pandas as pd
list=[81,45,67,88,97]
series=pd.Series(list)

print("List:",list,"\n","type:",type(list))
print("Series:",series,"\n","type:",type(series))

# Defining index of series
s1=pd.Series(list,index=["r1","r2","r3","r4","r5"])
print("Series with defined index:","\n",s1)

# Accessing elements in series
print(s1["r4"])    # with defined index

print(series[3])   # with default index
```

2) Data frames
a. Make a Pandas Dataframe with-
2-D python list, 
 Python Dictionary,
 List of lists,
 List of Dictionaries,


```python
import pandas as pd

# Data frame with 2-D lists
print("Data Frame with 2-D list:\n")
list_2d=[

    [1,2,3],
    [4,5,6],
    [7,8,9]
]
print(list_2d, type(list_2d))

df=pd.DataFrame(list_2d)
print(df)

df1=pd.DataFrame(list_2d,columns=["c1","c2","c3"],index=["r1","r2","r3"])
print(df1)

# Data frame with dictionary
print("Data Frame with dictionary:\n")
dic={
    "RollNo":[1,2,3,4],
    "Name":["Aryan","Riya","suman","ram"],
    "Marks":[82,88,85,90]
}

print(dic,type(dic))
df=pd.DataFrame(dic)
print(df)

df1=pd.DataFrame(dic,index=["s1","s2","s3","s4"])
print(df1)

# Data frame with lists of lists
print("Data Frame with lists of lists:\n")
list_of_list=[
    ["a","b"],["c"],["d","e","f"],["g","h"]
]
print(list_of_list,type(list_of_list))
df=pd.DataFrame(list_of_list)
print(df)

df1=pd.DataFrame(list_of_list,index=[1,2,3,4])
print(df1)

# Data frame with list of dictionaries
print("Data Frame with list of dictionaries:\n ")
list_of_dic=[
    {"names":["A","B","C","D"],
     "marks":[20,18,17,18]},

     {
         "class":["Ds","IT","DS","CSE"]
     }
]
print(list_of_dic,type(list_of_dic))

df=pd.DataFrame(list_of_dic)
print(df)

df1=pd.DataFrame(list_of_dic,index=[1,2])
print(df1)

```

3) Data iteration
a.Different ways to iterate over rows in pandas Data frame, b.Selecting rows on the basis of conditons, c.Select any row using iloc, d.limited rows selection with given column, e.drop rows based on conditon applied on column, f.Insert row at given position, g.Create list of rows in pandas

```python
import pandas as pd

d={
    'Name':['Aman',"Anika","Ram","Suman"],
    'Marks':[85,89,88,83],
    'City':["jaipur","ajmer","jaipur","kota"]
}
df=pd.DataFrame(d)
print("Original Data Frame:\n",df)

# Data Iteration
print("\nMarks of students:") 
for index, rows in df.iterrows():
  print(index,rows['Name'],rows['Marks'])

# Selecting rows based on condition
print("\nStudents with marks more than 85")
print(df[df['Marks']>85])

# Selecting rows using iloc
print("\nUsing iloc:\n")
print(df.iloc[0:2,0:2])

# Limited rows selection with given column
print("\nSelecting limited rows using cloumns")
print("\nName and City of Students")
print(df.loc[0:4,["Name","City"]]) 

# Drop Rows based on conditons
print("\nDropping Students based in jaipur city")
print(df[df['City']!='jaipur'])

# inserting rows at specified location
new_row=pd.DataFrame({
  'Name':['Neha'],
  'Marks':[84],
  'City':['ajmer']
})
df_new1=df.iloc[:3]
df_new2=df.iloc[3:]
print("\nNew DataFrame")
df_new=pd.concat([df_new1,new_row,df_new2]).reset_index(drop=True)
print(df_new)

# Create list of Data Frame
print("\nLists from Data frame:")
print(df_new.values.tolist())
```
