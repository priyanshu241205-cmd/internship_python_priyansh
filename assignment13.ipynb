1) Replace Nan with 0 and Interchange 3 rows and 3 columns of 2D array [[6, -8, 73, -110], [np.nan, -8, 0, 94]]

```python
# creating arrays
import numpy as np
a1=np.array([[6,-8,73,-110],[np.nan,-8,0,94]])

print(a1)

# replacing nan with 0
res=np.nan_to_num(a1,nan=0)
print(res)

# interchange 3 rows and 3 columns-- tranpose
print("\n",np.transpose(res))
```

3) Replace NaN values with average of columns

```python
import numpy as np
import math
a1=np.array([[6,-8,73,-110],[np.nan,-8,0,94]])

print(a1)

#finding the array of non nan values
non_nan=a1[~np.isnan(a1)]
print(non_nan)

# computing average of non nan values
avg=np.average(non_nan)
print(math.floor(avg))

# replacing avg with nan
res=np.nan_to_num(a1,nan=avg)
print(res)


```

4) Replace negative value with zero in numpy array using replace

```python
import numpy as np
a1=np.array([[6,-8,73,-110],[np.nan,-8,0,94]])

print(a1)

# method 1-- using np.where(array condition, value, arr name)
res=np.where(a1<0,0,a1)
print(res)

# method 2
a1[a1<0]=0
print(a1)
```

5) Study the following program import numpy as np # create a numpy 1d-arrays arr1 = np.array([3, 4]) arr2 = np.array([1, 0]) # find average of NumPy arrays avg = (arr1 + arr2) / 2 print("Average of NumPy arrays:\n", avg) -> Calculate average mean median mode values of two NumPy 2d-arrays 

```python
import numpy as np 
from scipy import stats
# create a numpy 1d-arrays
arr1 = np.array([3, 4])
arr2 = np.array([1, 0])
 # find average of NumPy arrays 
avg = (arr1 + arr2) / 2 
print("Average of NumPy arrays:\n", avg) 

combined=np.concatenate((arr1,arr2))
print(combined)

print("Mean:\n",np.mean(combined))
print("\nMedian:\n",np.median(combined))
mode=stats.mode(combined,keepdims=True)
print("\nMode:\n",mode.mode)
```

6) Solve the following equation using linalg() and inverse Matrix method x - 2y + 3z = 9 -x + 3y - z = -6 2x - 5y + 5z = 17

```python
# using linalg()
import numpy as np

arr1=np.array([[1,3,-1],[2,-5,5],[1,-2,3]])
print("Coefficient matrix:\n",arr1)

arr2=np.array([-6,17,9])
print("\nConstant Vector:\n",arr2)

res=np.linalg.solve(arr1,arr2)
print("\nResult:\n",res)


# using inverse-- np.dot()
arr_inv=np.linalg.inv(arr1)
print("\n",arr_inv)

res=np.dot(arr_inv,arr2)
print("\n",res)
```
