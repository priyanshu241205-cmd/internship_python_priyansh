```python
def max(a,b,c):
    return a if a > b and a > c else b if b > c and b > a else c
max(34,54,76)
```

Q2 Write a Python function that takes a list and returns a new list with distinct elements from the first list.

```python
def distinctelements(list1):
    return list(set(list1))
numbers = [1, 2, 2, 3, 4, 4, 4, 5]
print(distinctelements(numbers)) 
```

Q3 Write a Python function to multiply all the numbers in a list.

```python
def multiply(num):
    result = 1  
    for num in numbers:
        result *= num    
    return result
numbers = [1, 2, 3, 4, 5]    
multiply(numbers)
```

Q4 Write a Python function to calculate the factorial of a number (a non-negative integer). The function accepts the number as an argument.

```python
def calculate_factorial(n):
    if n < 0:
        print("Factorial is not defined for negative numbers.")
        return None
    result = 1
    while n > 1:
        result *= n
        n -= 1
    return result


print(calculate_factorial(5))
print(calculate_factorial(0))
```

Q5 Write a Python program to reverse a string.

```python
def reverse_string(s):
    return s[::-1]
str1 = "Hello, World!"
print(reverse_string(str1))
```

Q6 Write a Python function to check whether a number falls within a given range.

```python
def range(number, start, end):
    return start <= number <= end

print(range(5, 1, 10))
print(range(12, 1, 10))
print(range(1, 1, 10))
```

Q7 Write a Python function to Print Even Numbers from a Given List

```python
def even_odd(num):
    for i in num:   
        if i % 2 == 0:
            print(f"{i} is even.")
        else:
            print(f"{i} is odd.")

list = [10, 15, 20, 25, 30]
even_odd(list)      
```

Q8 Write a Python function that takes a number as a parameter and checks whether the number is prime or not.

```python
def prime(number):
   if number <= 1:
       return False
   if number == 2:
       return True
   if number % 2 == 0:
       return False
   i = 3
   while i * i <= number:
       if number % i == 0:
           return False
       i += 2           
   return True
print(prime(11))

```

Q9 Write a Python function that accepts a string and counts the number of upper and lower case letters. 

```python
def count_case_short(text):
    upper = sum(1 for char in text if char.isupper())
    lower = sum(1 for char in text if char.islower())
    return {"upper": upper, "lower": lower}

# Example usage:
print(count_case_short("Python 3.12"))
```
