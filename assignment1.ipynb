Q1 Practice print statement: - with multiple object - with seperator - with "ends"

```python
name = "Alice"
age = 25

print("Hello", name, "you are", age, "years old.")


# Separate with a hyphen
print("2026", "05", "31", sep="-")

# Separate with a forward slash
print("home", "user", "documents", sep="/")

# Separate with a newline character (\n)
print("Apple", "Banana", "Cherry", sep="\n")




# Keep the next print on the same line with a space
print("Hello", end=" ")
print("World!")

# Connect items with a series of dots
print("Loading", end="...")
print(" Done!")
```
