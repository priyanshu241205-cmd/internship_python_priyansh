Q1 Replace Nan with 0 and Interchange 3 rows and 3 columns of 2D array [[6, -8, 73, -110], [np.nan, -8, 0, 94]] 

```python
import numpy as np

arr = np.array([[6, -8, 73, -110],
                [np.nan, -8, 0, 94]])

print("Original Array:")
print(arr)

# Replace NaN with 0
arr = np.nan_to_num(arr, nan=0)

print("\nAfter Replacing NaN with 0:")
print(arr)

# Interchange rows and columns
transpose_arr = arr.T

print("\nTranspose Array:")
print(transpose_arr)
```

Q2 Move axes of 3D array to new positions

```python
import numpy as np

arr = np.arange(24).reshape(2, 3, 4)

print("Original Shape:")
print(arr.shape)

# Move axis 0 to position 2
new_arr = np.moveaxis(arr, 0, 2)

print("\nNew Shape:")
print(new_arr.shape)

print("\nArray after moving axes:")
print(new_arr)
```

Q3 Replace NaN values with average of columns 

```python
import numpy as np

arr = np.array([[10, np.nan, 30],
                [20, 40, np.nan],
                [30, 50, 60]])

print("Original Array:")
print(arr)

# Column means
col_mean = np.nanmean(arr, axis=0)

# Find NaN positions
inds = np.where(np.isnan(arr))

# Replace NaN with column mean
arr[inds] = np.take(col_mean, inds[1])

print("\nAfter Replacing NaN with Column Mean:")
print(arr)
```

Q4 Replace negative value with zero in numpy array using replace

```python
import numpy as np

arr = np.array([10, -5, 20, -8, 15, -3])

print("Original Array:")
print(arr)

arr[arr < 0] = 0

print("\nAfter Replacing Negative Values with 0:")
print(arr)
```
